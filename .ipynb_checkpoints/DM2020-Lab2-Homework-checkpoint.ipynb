````python
import pandas as pd

### training data
anger_train = pd.read_csv("data/semeval/train/anger-ratings-0to1.train.txt",
                         sep="\t", header=None,names=["id", "text", "emotion", "intensity"])
sadness_train = pd.read_csv("data/semeval/train/sadness-ratings-0to1.train.txt",
                         sep="\t", header=None, names=["id", "text", "emotion", "intensity"])
fear_train = pd.read_csv("data/semeval/train/fear-ratings-0to1.train.txt",
                         sep="\t", header=None, names=["id", "text", "emotion", "intensity"])
joy_train = pd.read_csv("data/semeval/train/joy-ratings-0to1.train.txt",
                         sep="\t", header=None, names=["id", "text", "emotion", "intensity"])

### testing data
anger_test = pd.read_csv("data/semeval/dev/anger-ratings-0to1.dev.gold.txt",
                         sep="\t", header=None, names=["id", "text", "emotion", "intensity"])
sadness_test = pd.read_csv("data/semeval/dev/sadness-ratings-0to1.dev.gold.txt",
                         sep="\t", header=None, names=["id", "text", "emotion", "intensity"])
fear_test = pd.read_csv("data/semeval/dev/fear-ratings-0to1.dev.gold.txt",
                         sep="\t", header=None, names=["id", "text", "emotion", "intensity"])
joy_test = pd.read_csv("data/semeval/dev/joy-ratings-0to1.dev.gold.txt",
                         sep="\t", header=None, names=["id", "text", "emotion", "intensity"])

# combine 4 sub-dataset
train_df = pd.concat([anger_train, fear_train, joy_train, sadness_train], ignore_index=True)
test_df = pd.concat([anger_test, fear_test, joy_test, sadness_test], ignore_index=True)

# shuffle dataset
train_df = train_df.sample(frac=1)
test_df = test_df.sample(frac=1)
````

### Prepare the training and testing data

````python
from sklearn.feature_extraction.text import CountVectorizer

# build analyzers (bag-of-words)
BOW_500 = CountVectorizer(max_features=500, tokenizer=nltk.word_tokenize) 

# apply analyzer to training data
BOW_500.fit(train_df['text'])

# for a classificaiton problem, you need to provide both training & testing data
X_train = BOW_500.transform(train_df['text'])
y_train = train_df['emotion']

X_test = BOW_500.transform(test_df['text'])
y_test = test_df['emotion']

## take a look at data dimension is a good habbit
print('X_train.shape: ', X_train.shape)
print('y_train.shape: ', y_train.shape)
print('X_test.shape: ', X_test.shape)
print('y_test.shape: ', y_test.shape)
````

### ** >>> Exercise 4 (Take home): **  
Build a model using a ```Naive Bayes``` model and train it. What are the testing results?

````python
# Answer here
from sklearn.naive_bayes import MultinomialNB

NB_model = MultinomialNB()

# train
NB_model = NB_model.fit(X_train, y_train)

# prediction
y_train_pred = NB_model.predict(X_train)
y_test_pred = NB_model.predict(X_test)

# accurancy
from sklearn.metrics import accuracy_score

acc_train = accuracy_score(y_true=y_train, y_pred=y_train_pred)
acc_test = accuracy_score(y_true=y_test, y_pred=y_test_pred)

print('training accuracy: {}'.format(round(acc_train, 2)))
print('testing accuracy: {}'.format(round(acc_test, 2)))
````

````python
# Answer here
# Using glove_twitter_25_model
word_list = ['happy', 'angry', 'sad', 'fear']

topn = 15 # print out top 15 words
happy_words = ['happy'] + [word_ for word_, sim_ in glove_twitter_25_model.most_similar('happy', topn=topn)]
angry_words = ['angry'] + [word_ for word_, sim_ in glove_twitter_25_model.most_similar('angry', topn=topn)]        
sad_words = ['sad'] + [word_ for word_, sim_ in glove_twitter_25_model.most_similar('sad', topn=topn)]        
fear_words = ['fear'] + [word_ for word_, sim_ in glove_twitter_25_model.most_similar('fear', topn=topn)]        

print('happy_words: ', happy_words)
print('angry_words: ', angry_words)
print('sad_words: ', sad_words)
print('fear_words: ', fear_words)

target_words = happy_words + angry_words + sad_words + fear_words
cn = topn + 1
color = ['b'] * cn + ['g'] * cn + ['r'] * cn + ['y'] * cn

import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

## prepare training word vectors
size = 100  # only use 100 here
target_size = len(target_words)
all_word = list(glove_twitter_25_model.vocab.keys())
word_train = target_words + all_word[:size]
X_train = glove_twitter_25_model[word_train]

## t-SNE model
tsne = TSNE(n_components=2, metric='cosine', random_state=28)

## training
X_tsne = tsne.fit_transform(X_train)

## plot the result
plt.figure(figsize=(7.5, 7.5), dpi=115)
plt.scatter(X_tsne[:target_size, 0], X_tsne[:target_size, 1], c=color)
for label, x, y in zip(target_words, X_tsne[:target_size, 0], X_tsne[:target_size, 1]):
    plt.annotate(label, xy=(x,y), xytext=(0,0),  textcoords='offset points')
plt.show()
````

````python
# ONLY First time, convert json data to csv with the data we care and split into train and test
import json
import pandas as pd

# take the _source and tweet data for json
data = []
with open('tweets_DM.json') as f:
    for line in f:
        data.append(json.loads(line)['_source']['tweet'])
        
tweet = pd.DataFrame.from_dict(data)
identification = pd.read_csv('./data_identification.csv')

print('identification shape: ', identification.shape)

# merge two datas
combine = pd.merge(tweet, identification, on='tweet_id')

# extract train and test 
Train_tweet = combine.loc[combine['identification']=='train']
Test_tweet = combine.loc[combine['identification']=='test']

# drop the identification label
Train_tweet = Train_tweet.drop(['identification'], axis=1)
Test_tweet = Test_tweet.drop(['identification'], axis=1)
print('Train: ', Train_tweet.shape)
print('Test: ', Test_tweet.shape)

# combine emotion(the label) also
emo = pd.read_csv('./emotion.csv', engine='python')
print('Emotion data length: ', emo.shape)
Train_tweet = pd.merge(Train_tweet, emo, on='tweet_id')

# store it for next time usage

# use newline here in case Windows add '\r\n' as the end of the line
with open('tweet_train.csv', mode='w', newline='\n', encoding='utf-8') as f:
    Train_tweet.to_csv(f, index=False, line_terminator='\r\n')
with open('tweet_test.csv', mode='w', newline='\n', encoding='utf-8') as f:
    Test_tweet.to_csv(f, index=False)
````

````python
import pandas as pd
import numpy as np
import nltk
import keras
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import auc, f1_score
````

````python
Train_tweet = pd.read_csv('tweet_train.csv', lineterminator='\n')
Train_tweet.head()
````

````python
#Train_tweet = pd.read_csv('tweet_train.csv', lineterminator='\n')
Train_data = Train_tweet.replace({r'\s+$': '', r'^\s+': ''}, regex=True).replace(r'\n',  ' ', regex=True)
Train_data.rename(columns = {'emotion\r':'emotion'}, inplace=True)
Train_data.head()
````

````python
Test_tweet = pd.read_csv('tweet_test.csv', lineterminator='\n')
Test_data = Test_tweet.replace({r'\s+$': '', r'^\s+': ''}, regex=True).replace(r'\n',  ' ', regex=True)
Test_data.rename(columns = {'text\r':'text'}, inplace=True)
Test_data.head()
````

````python
Train_data = Train_data.sample(frac=1)
Train_data.head()
````

````python
Test_data = Test_data.sample(frac=1)
Test_data.head()
````

````python
import re
#replace some special text by regularize words
emailsRegex=re.compile(r'[\w\.-]+@[\w\.-]+')
userMentionsRegex=re.compile(r'(?<=^|(?<=[^a-zA-Z0-9-_\.]))@([A-Za-z]+[A-Za-z0-9]+)')
urlsRegex=re.compile('r(f|ht)(tp)(s?)(://)(.*)[.|/][^ ]+')
numsRegex=re.compile(r"\b\d+\b")
````

````python
import time
texts = Train_data['text']
start = time.time()
texts = texts.apply(lambda a:a.replace("<LH>",""))
texts = texts.apply(lambda a:userMentionsRegex.sub(' USER',a))
texts = texts.apply(lambda a:emailsRegex.sub(' EMAIL',a))
texts = texts.apply(lambda a:urlsRegex.sub(' URL',a))
texts = texts.apply(lambda a:numsRegex.sub(' NUM',a))
print("TIME: %f",time.time()-start)
````

````python
# remove stopword
from nltk.corpus import stopwords
stop = stopwords.words('english')

texts.apply(lambda x: ' '.join([word for word in x.split() if word not in stop]))
#texts.str.lower()
````

````python
from nltk.stem import SnowballStemmer 
snowball_stemmer = SnowballStemmer('english')

#snowball_stemmer.stem(w) for w in w_tokenizer.tokenize(str(text))
texts.apply(lambda x: ' '.join([snowball_stemmer.stem(word) for word in x.split()]))
````

````python
#texts = list(Train_data['text'])
#labels = list(Train_data['emotion'])
#texts
Train_data['text'] = texts 
Train_data['text'].head()
````

````python
# Label encoder
L_ENC = LabelEncoder()
LABEL = L_ENC.fit_transform(Train_data['emotion'])

import keras
LABEL_category = keras.utils.to_categorical(LABEL)  # convert to one hot encoding

# labels
LABEL_category[0:5]
````

````python
# TFIDF
TFIDF = TfidfVectorizer(max_features=7000, tokenizer=nltk.word_tokenize)

TRAIN_TFIDF = TFIDF.fit_transform(Train_data['text'])
print(TRAIN_TFIDF.shape)
````

````python
X_train, X_val, Y_train, Y_val = train_test_split(TRAIN_TFIDF, LABEL_category,  test_size=0.25) #split the data
print('X_train.shape: ', X_train.shape)
print('y_train.shape: ', Y_train.shape)
print('X_test.shape: ', X_val.shape)
print('y_test.shape: ', Y_val.shape)
````

````python
# I/O check
input_shape = X_train.shape[1]
print('input_shape: ', input_shape)

output_shape = len(L_ENC.classes_)
print('output_shape: ', output_shape)
````

````python
# model - NN
from keras.models import Model
from keras.layers import Input, Dense
from keras.layers import ReLU, Softmax

# input layer
model_input = Input(shape=(input_shape, ))  # max_features
X = model_input

# 1st hidden layer
X_W1 = Dense(units=64)(X)  # 64
H1 = ReLU()(X_W1)

# 2nd hidden layer
H1_W2 = Dense(units=64)(H1)  # 64
H2 = ReLU()(H1_W2)

# 3nd hidden layer
H2_W3 = Dense(units=32)(H2)  # 64
H3 = ReLU()(H2_W3)

# output layer
H3_W4 = Dense(units=output_shape)(H3)  # 8
H4 = Softmax()(H3_W4)

model_output = H4

# create model
model = Model(inputs=[model_input], outputs=[model_output])

# loss function & optimizer
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
````

### Train NN

````python
# training setting
epochs = 15
batch_size = 32

# training!
history = model.fit(X_train, Y_train, 
                    epochs=epochs, 
                    batch_size=batch_size,
                    validation_data = (X_val, Y_val))
print('training finish')
````

````python
#reasonable epochs is 
# retrain on whole dataset
epochs = 3
batch_size = 32

history = model.fit(TRAIN_TFIDF, LABEL_category,
                    epochs=epochs, 
                    batch_size=batch_size)
print('training finish')
````

````python
# Predict test result
TEST_TFIDF = TFIDF.transform(Test_data['text'])

# predict
pred_result = model.predict(TEST_TFIDF, batch_size=32)
pred_label = np.argmax(pred_result, axis=1)

# convert label back to original result
test_label = L_ENC.inverse_transform(pred_label)
print(test_label[0:10])
print(test_label.shape)
````

````python
# concat to result and save it
RESULT = pd.DataFrame(columns=['id','emotion'])
RESULT['id'] = Test_data['tweet_id']
RESULT['emotion'] = test_label
print(RESULT.shape)

# save it for submission
RESULT.to_csv('result.csv', index=False)
````
